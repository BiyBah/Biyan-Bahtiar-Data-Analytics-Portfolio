# Import library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# %pip install faker
from faker import Faker
from datetime import datetime
```

# Initialize Faker

```python
fake_id = Faker('id_ID')

Faker.seed(0)
np.random.seed(0)
```

## Table country

```python
country = pd.DataFrame()
n_country = 1

country_key = [10001+i for i in range(n_country)]

country['country_id'] = country_key
country['country'] = ['Indonesia']

country
```

## Table mfg_year

```python
mfg_year = pd.DataFrame()
n_mfg_year = 4

mfg_year_key = [20001+i for i in range(n_mfg_year)]

mfg_year['mfg_year_id'] = mfg_year_key
mfg_year['mfg_year'] = [2010, 2011, 2012, 2013]

mfg_year
mfg_year
```

## Table transmission

```python
transmission = pd.DataFrame()
n_transmission = 2

transmission_key = [30001+i for i in range(n_transmission)]

transmission['transmission_id'] = transmission_key
transmission['transmission'] = ['Automatic', 'Manual']

transmission
```

## Table body

```python
body = pd.DataFrame()
n_body = 5

body_key = [40001+i for i in range(n_body)]
body_list = [
    "Sedan",
    "Hatchback",
    "SUV",
    "MPV",
    "Van"
]

body['body_id'] = body_key
body['body'] = body_list

body
```

## Table brand

```python
brand = pd.DataFrame()
n_brand = 10

brand_list = [
    "Toyota",
    "Ford",
    "BMW",
    "Mercedes-Benz",
    "Honda",
    "Chevrolet",
    "Audi",
    "Hyundai",
    "Nissan",
    "Volkswagen"
]
brand_key = [50001+i for i in range(n_brand)]

brand['brand_id'] = brand_key
brand['brand'] = brand_list

brand
```

## Table city, province and users

```python
city = pd.DataFrame()
province = pd.DataFrame()
users = pd.DataFrame()

n_city = 40
n_province = 6
n_users = 100

city_key = [60001+i for i in range(n_city)]
province_key = [70001+i for i in range(n_province)]
users_key = [80001+i for i in range(n_users)]

# table city
city['city_id'] = city_key
city['city'] = [fake_id.unique.city() for i in range(n_city)]
city['province_id'] = [np.random.choice(np.array(province_key)) for i in range(n_city)]
city.drop_duplicates(subset='city', inplace=True)

display(city)

# table province
province['province_id'] = province_key
province['province'] = [fake_id.unique.state() for i in range(n_province)]
province['country_id'] = [10001 for i in range(n_province)]
province.drop_duplicates(subset='province', inplace=True)

display(province)

# table users
users['user_id'] = users_key
users['user_name'] = [fake_id.unique.name() for i in range(n_users)]
users['password'] = [fake_id.password(length=8, special_chars=True,digits=True, upper_case=True,lower_case=True) for i in range(n_users)]
users['phone_number'] = [fake_id.phone_number() for i in range(n_users)]
users['social_media_address'] = [''.join(['instagram.com/', i.lower().replace('.','').replace(' ','_').replace(',','')]) for i in users['user_name']]
users['domicile_address'] = [fake_id.address() for i in range(n_users)]
users['domicile_address'] = users['domicile_address'].str.replace('\n',',')
users['city_id'] = [np.random.choice(np.array(city_key)) for i in range(n_users)]
users['date_created'] = [fake_id.date_between()]

display(users)
```

## Table ads

```python
ads = pd.DataFrame()
n_ads = 70

ads_key = [90001+i for i in range(n_ads)]

ads['ads_id'] = ads_key
ads['title'] = [fake_id.text()[:10] for i in range(n_ads)]
ads['description'] = [fake_id.text() if len(fake_id.text()) <= 30 else fake_id.text()[:30] for i in range(n_ads)]
ads['user_id'] = [np.random.choice(np.array(users_key)) for i in range(n_ads)]

display(ads)
```

## Table bid, car and model

```python
model = pd.DataFrame()
car = pd.DataFrame()
bid = pd.DataFrame()

n_model = 40
n_car = 70
n_bid = 30

model_key = [11001+i for i in range(n_model)]
car_key = [12001+i for i in range(n_car)]
model_list = [
    "Toyota Corolla",
    "Honda Civic",
    "Ford F-150",
    "Chevrolet Silverado",
    "BMW 3 Series",
    "Mercedes-Benz C-Class",
    "Audi A4",
    "Tesla Model 3",
    "Honda Accord",
    "Nissan Altima",
    "Volkswagen Golf",
    "Toyota Camry",
    "Hyundai Elantra",
    "Jeep Grand Cherokee",
    "Mazda CX-5",
    "Ford Mustang",
    "Chevrolet Tahoe",
    "Subaru Outback",
    "Kia Sorento",
    "Honda CR-V",
    "Toyota RAV4",
    "Hyundai Tucson",
    "Nissan Rogue",
    "Chevrolet Equinox",
    "Ford Explorer",
    "Dodge Charger",
    "Jeep Wrangler",
    "BMW X5",
    "Tesla Model S",
    "Porsche 911",
    "Chevrolet Malibu",
    "Honda Pilot",
    "Hyundai Santa Fe",
    "Lexus RX",
    "Mercedes-Benz E-Class",
    "Nissan Sentra",
    "Volkswagen Passat",
    "Toyota Highlander",
    "Ram 1500",
    "Ford Escape"
]

model['model_id'] = model_key
model['model_name'] = model_list
model['transmission_id'] = [np.random.choice(np.array(transmission_key)) for i in range(n_model)]
model['body_id'] = [np.random.choice(np.array(body_key)) for i in range(n_model)]
model['brand_id'] = [np.random.choice(np.array(brand_key)) for i in range(n_model)]

display(model)

car['car_id'] = car_key
car['permit_bid'] = [fake_id.boolean(chance_of_getting_true = 50) for i in range(n_car)]
car['color'] = [fake_id.color_name() for i in range(n_car)]
car['mileage_km'] = [np.random.randint(10,100)*10000 for i in range(n_car)]
car['mfg_year_id'] = [np.random.choice(np.array(mfg_year_key)) for i in range(n_car)]
car['model_id'] = [np.random.choice(np.array(model_key)) for i in range(n_car)]
car['ads_id'] = [np.random.choice(np.array(ads_key)) for i in range(n_car)]

display(car)

car_bid_key = list(car[car['permit_bid']==True]['car_id'].unique())
bid['bid_id'] = [13001+i for i in range(n_bid)]
bid['bid_price'] = [np.random.randint(10,100)*10000000 for i in range(n_bid)]
bid['user_id'] = [np.random.choice(np.array(users_key)) for i in range(n_bid)]
bid['car_id'] = [np.random.choice(np.array(car_bid_key)) for i in range(n_bid)]

display(bid)
```
